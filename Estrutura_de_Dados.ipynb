## 2. Árvore Binária de Busca (BST)

### 2.1. Explicação da Lógica
A Árvore Binária de Busca (BST) é uma estrutura de dados onde cada nó possui no máximo dois filhos. A inserção, remoção e busca são realizadas de forma recursiva, seguindo a propriedade da BST: valores menores que o nó atual são armazenados à esquerda, e valores maiores à direita.

### 2.2. Complexidade
- **Inserção**: O(h), onde h é a altura da árvore. No pior caso (árvore desbalanceada), h = n, resultando em O(n). No melhor caso (árvore balanceada), h = log(n), resultando em O(log n).
- **Busca**: O(h), semelhante à inserção.
- **Remoção**: O(h), pois envolve a busca do nó a ser removido e possíveis ajustes na árvore.
- **Altura**: O(n), pois todos os nós são visitados.

### 2.3. Resultados de Execução  
| Tamanho dos Dados | Tempo de Inserção (s) | Tempo de Busca (s) | Tempo de Remoção (s) | Tempo de Altura (s) |
|-------------------|-----------------------|--------------------|----------------------|---------------------|
| 100               | 0.000358              | 0.000010           | 0.000011             | 0.000087            |
| 1.000             | 0.012658              | 0.000009           | 0.000012             | 0.000551            |
| 10.000            | 0.127250              | 0.000011           | 0.000009             | 0.006257            |
| 100.000           | 1.205206              | 0.000020           | 0.000010             | 0.052168            |

### 2.4. Análise
Os tempos de execução aumentam conforme o tamanho dos dados, mas a BST mantém uma eficiência razoável para operações de inserção, busca e remoção. A altura da árvore é pequena, o que garante operações rápidas.

## Questão 1: Implementação de uma Árvore Binária de Busca (BST)


```python
class Node:
    def __init__(self, value):
        self.value = value  # Pode ser um valor simples ou uma tupla (codigo_produto, quantidade)
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def inserir(self, value):
        if self.root is None:
            self.root = Node(value)
        else:
            self._inserir_recursivo(self.root, value)

    def _inserir_recursivo(self, node, value):
        # Se o valor for uma tupla, comparamos apenas o primeiro elemento (codigo_produto)
        valor_comparacao = value[0] if isinstance(value, tuple) else value
        node_valor_comparacao = node.value[0] if isinstance(node.value, tuple) else node.value

        if valor_comparacao < node_valor_comparacao:
            if node.left is None:
                node.left = Node(value)
            else:
                self._inserir_recursivo(node.left, value)
        elif valor_comparacao > node_valor_comparacao:
            if node.right is None:
                node.right = Node(value)
            else:
                self._inserir_recursivo(node.right, value)

    def remover(self, value):
        # Se o valor for uma tupla, comparamos apenas o primeiro elemento (codigo_produto)
        valor_comparacao = value[0] if isinstance(value, tuple) else value
        self.root = self._remover_recursivo(self.root, valor_comparacao)

    def _remover_recursivo(self, node, value):
        if node is None:
            return node

        # Se o valor do nó for uma tupla, comparamos apenas o primeiro elemento (codigo_produto)
        node_valor_comparacao = node.value[0] if isinstance(node.value, tuple) else node.value

        if value < node_valor_comparacao:
            node.left = self._remover_recursivo(node.left, value)
        elif value > node_valor_comparacao:
            node.right = self._remover_recursivo(node.right, value)
        else:
            if node.left is None:
                return node.right
            elif node.right is None:
                return node.left
            temp = self._minimo(node.right)
            node.value = temp.value
            node.right = self._remover_recursivo(node.right, temp.value[0] if isinstance(temp.value, tuple) else temp.value)
        return node

    def altura(self):
        return self._altura_recursivo(self.root)

    def _altura_recursivo(self, node):
        if node is None:
            return -1
        return max(self._altura_recursivo(node.left), self._altura_recursivo(node.right)) + 1

    def buscar(self, value):
        # Se o valor for uma tupla, comparamos apenas o primeiro elemento (codigo_produto)
        valor_comparacao = value[0] if isinstance(value, tuple) else value
        return self._buscar_recursivo(self.root, valor_comparacao)

    def _buscar_recursivo(self, node, value):
        if node is None:
            return None

        # Se o valor do nó for uma tupla, comparamos apenas o primeiro elemento (codigo_produto)
        node_valor_comparacao = node.value[0] if isinstance(node.value, tuple) else node.value

        if value == node_valor_comparacao:
            return node
        elif value < node_valor_comparacao:
            return self._buscar_recursivo(node.left, value)
        else:
            return self._buscar_recursivo(node.right, value)

    def maximo(self):
        return self._maximo(self.root).value if self.root else None

    def _maximo(self, node):
        while node.right is not None:
            node = node.right
        return node

    def minimo(self):
        return self._minimo(self.root).value if self.root else None

    def _minimo(self, node):
        while node.left is not None:
            node = node.left
        return node

    def sucessor(self, value):
        return self._sucessor_recursivo(self.root, value)

    def _sucessor_recursivo(self, node, value):
        if node is None:
            return None
        if node.value[0] <= value:
            return self._sucessor_recursivo(node.right, value)
        else:
            left = self._sucessor_recursivo(node.left, value)
            return left if left is not None else node.value

    def predecessor(self, value):
        return self._predecessor_recursivo(self.root, value)

    def _predecessor_recursivo(self, node, value):
        if node is None:
            return None
        if node.value[0] >= value:
            return self._predecessor_recursivo(node.left, value)
        else:
            right = self._predecessor_recursivo(node.right, value)
            return right if right is not None else node.value

    def percurso_pre_ordem(self):
        self._percurso_pre_ordem(self.root)

    def _percurso_pre_ordem(self, node):
        if node:
            print(node.value, end=" ")
            self._percurso_pre_ordem(node.left)
            self._percurso_pre_ordem(node.right)

    def percurso_pos_ordem(self):
        self._percurso_pos_ordem(self.root)

    def _percurso_pos_ordem(self, node):
        if node:
            self._percurso_pos_ordem(node.left)
            self._percurso_pos_ordem(node.right)
            print(node.value, end=" ")

    def percurso_em_ordem(self):
        self._percurso_em_ordem(self.root)

    def _percurso_em_ordem(self, node):
        if node:
            self._percurso_em_ordem(node.left)
            print(node.value, end=" ")
            self._percurso_em_ordem(node.right)

    def eh_balanceada(self):
        return self._eh_balanceada_recursivo(self.root)

    def _eh_balanceada_recursivo(self, node):
        if node is None:
            return True
        altura_esquerda = self._altura_recursivo(node.left)
        altura_direita = self._altura_recursivo(node.right)
        if abs(altura_esquerda - altura_direita) <= 1 and self._eh_balanceada_recursivo(node.left) and self._eh_balanceada_recursivo(node.right):
            return True
        return False

    def valores_no_intervalo(self, inicio, fim):
        valores = []
        self._valores_no_intervalo_recursivo(self.root, inicio, fim, valores)
        return valores

    def _valores_no_intervalo_recursivo(self, node, inicio, fim, valores):
        if node is None:
            return
        if inicio < node.value[0]:
            self._valores_no_intervalo_recursivo(node.left, inicio, fim, valores)
        if inicio <= node.value[0] <= fim:
            valores.append(node.value)
        if fim > node.value[0]:
            self._valores_no_intervalo_recursivo(node.right, inicio, fim, valores)
```

## Questão 3: Sistema de Gerenciamento de Inventário usando BST


```python
class SistemaInventario:
    def __init__(self):
        self.bst = BST()

    def inserir(self, codigo_produto, quantidade):
        if self.bst.buscar(codigo_produto):
            self.atualizar_quantidade(codigo_produto, quantidade)
        else:
            self.bst.inserir((codigo_produto, quantidade))

    def remover(self, codigo_produto):
        self.bst.remover(codigo_produto)

    def buscar(self, codigo_produto):
        node = self.bst.buscar(codigo_produto)
        return node.value if node else None

    def atualizar_quantidade(self, codigo_produto, nova_quantidade):
        self.remover(codigo_produto)
        self.inserir(codigo_produto, nova_quantidade)

    def minimo(self):
        return self.bst.minimo()

    def maximo(self):
        return self.bst.maximo()

    def produtos_acima_quantidade(self, quantidade_minima):
        produtos = []
        self._produtos_acima_quantidade_recursivo(self.bst.root, quantidade_minima, produtos)
        return produtos

    def _produtos_acima_quantidade_recursivo(self, node, quantidade_minima, produtos):
        if node:
            if node.value[1] > quantidade_minima:
                produtos.append(node.value)
            self._produtos_acima_quantidade_recursivo(node.left, quantidade_minima, produtos)
            self._produtos_acima_quantidade_recursivo(node.right, quantidade_minima, produtos)
```

## Questão 5: Implementação de uma Árvore B


```python
class BTreeNode:
    def __init__(self, leaf=True):
        self.leaf = leaf
        self.keys = []
        self.children = []

class BTree:
    def __init__(self, t):
        self.root = BTreeNode()
        self.t = t  # Ordem da árvore B

    def inserir(self, key):
        root = self.root
        if len(root.keys) == (2 * self.t) - 1:
            new_root = BTreeNode(leaf=False)
            new_root.children.append(self.root)
            self._dividir_no(new_root, 0)
            self.root = new_root
        self._inserir_nao_cheio(self.root, key)

    def _inserir_nao_cheio(self, node, key):
        i = len(node.keys) - 1
        if node.leaf:
            node.keys.append(None)
            while i >= 0 and key < node.keys[i]:
                node.keys[i + 1] = node.keys[i]
                i -= 1
            node.keys[i + 1] = key
        else:
            while i >= 0 and key < node.keys[i]:
                i -= 1
            i += 1
            if len(node.children[i].keys) == (2 * self.t) - 1:
                self._dividir_no(node, i)
                if key > node.keys[i]:
                    i += 1
            self._inserir_nao_cheio(node.children[i], key)

    def _dividir_no(self, node, i):
        t = self.t
        y = node.children[i]
        z = BTreeNode(leaf=y.leaf)
        node.children.insert(i + 1, z)
        node.keys.insert(i, y.keys[t - 1])
        z.keys = y.keys[t: (2 * t) - 1]
        y.keys = y.keys[0: t - 1]
        if not y.leaf:
            z.children = y.children[t: 2 * t]
            y.children = y.children[0: t]

    def buscar(self, key):
        return self._buscar_recursivo(self.root, key)

    def _buscar_recursivo(self, node, key):
        i = 0
        while i < len(node.keys) and key > node.keys[i]:
            i += 1
        if i < len(node.keys) and key == node.keys[i]:
            return True
        if node.leaf:
            return False
        return self._buscar_recursivo(node.children[i], key)

    def remover(self, key):
        self._remover_recursivo(self.root, key)
        if len(self.root.keys) == 0 and not self.root.leaf:
            self.root = self.root.children[0]

    def _remover_recursivo(self, node, key):
        i = 0
        while i < len(node.keys) and key > node.keys[i]:
            i += 1
        if node.leaf:
            if i < len(node.keys) and key == node.keys[i]:
                node.keys.pop(i)
            return
        if i < len(node.keys) and key == node.keys[i]:
            self._remover_chave_interna(node, i)
            return
        if len(node.children[i].keys) < self.t:
            self._preencher_filho(node, i)
        if i > len(node.keys):
            self._remover_recursivo(node.children[i - 1], key)
        else:
            self._remover_recursivo(node.children[i], key)

    def _remover_chave_interna(self, node, i):
        t = self.t
        if len(node.children[i].keys) >= t:
            predecessor = self._maximo(node.children[i])
            node.keys[i] = predecessor
            self._remover_recursivo(node.children[i], predecessor)
        elif len(node.children[i + 1].keys) >= t:
            sucessor = self._minimo(node.children[i + 1])
            node.keys[i] = sucessor
            self._remover_recursivo(node.children[i + 1], sucessor)
        else:
            self._merge(node, i)
            self._remover_recursivo(node.children[i], node.keys[i])
            node.keys.pop(i)

    def _preencher_filho(self, node, i):
        t = self.t
        if i != 0 and len(node.children[i - 1].keys) >= t:
            self._emprestar_do_anterior(node, i)
        elif i != len(node.keys) and len(node.children[i + 1].keys) >= t:
            self._emprestar_do_proximo(node, i)
        else:
            if i != len(node.keys):
                self._merge(node, i)
            else:
                self._merge(node, i - 1)

    def _emprestar_do_anterior(self, node, i):
        filho = node.children[i]
        irmao = node.children[i - 1]
        filho.keys.insert(0, node.keys[i - 1])
        if not filho.leaf:
            filho.children.insert(0, irmao.children.pop())
        node.keys[i - 1] = irmao.keys.pop()

    def _emprestar_do_proximo(self, node, i):
        filho = node.children[i]
        irmao = node.children[i + 1]
        filho.keys.append(node.keys[i])
        if not filho.leaf:
            filho.children.append(irmao.children.pop(0))
        node.keys[i] = irmao.keys.pop(0)

    def _merge(self, node, i):
        t = self.t
        filho = node.children[i]
        irmao = node.children[i + 1]
        filho.keys.append(node.keys.pop(i))
        filho.keys.extend(irmao.keys)
        if not filho.leaf:
            filho.children.extend(irmao.children)
        node.children.pop(i + 1)

    def _maximo(self, node):
        while not node.leaf:
            node = node.children[-1]
        return node.keys[-1]

    def _minimo(self, node):
        while not node.leaf:
            node = node.children[0]
        return node.keys[0]

    def altura(self):
        return self._altura_recursivo(self.root)

    def _altura_recursivo(self, node):
        if node.leaf:
            return 0
        return 1 + self._altura_recursivo(node.children[0])

    def imprimir_em_ordem(self):
        self._imprimir_em_ordem_recursivo(self.root)

    def _imprimir_em_ordem_recursivo(self, node):
        if node:
            for i in range(len(node.keys)):
                if not node.leaf:
                    self._imprimir_em_ordem_recursivo(node.children[i])
                print(node.keys[i], end=" ")
            if not node.leaf:
                self._imprimir_em_ordem_recursivo(node.children[-1])

    def contar_chaves(self):
        return self._contar_chaves_recursivo(self.root)

    def _contar_chaves_recursivo(self, node):
        count = len(node.keys)
        if not node.leaf:
            for child in node.children:
                count += self._contar_chaves_recursivo(child)
        return count
```

## Questão 7: Implementação de um Dicionário em Python (MyDict)


```python
class MyDict:
    def __init__(self):
        self.dict = {}

    def inserir(self, chave, valor):
        self.dict[chave] = valor

    def remover(self, chave):
        if chave in self.dict:
            del self.dict[chave]

    def buscar(self, chave):
        return self.dict.get(chave, None)

    def atualizar(self, chave, valor):
        if chave in self.dict:
            self.dict[chave] = valor

    def tamanho(self):
        return len(self.dict)

    def inverter(self):
        return {v: k for k, v in self.dict.items()}

    def verificar_igualdade(self, outra_lista):
        return self.dict == outra_lista
```

## Questão 9: Sistema de Cadastro de Produtos usando Tabela Hash


```python
class SistemaCadastroProdutos:
    def __init__(self):
        self.tabela = {}

    def cadastrar_produto(self, codigo_produto, nome_produto):
        self.tabela[codigo_produto] = nome_produto

    def buscar_produto(self, codigo_produto):
        return self.tabela.get(codigo_produto, None)

    def verificar_produto(self, codigo_produto):
        return codigo_produto in self.tabela

    def excluir_produto(self, codigo_produto):
        if codigo_produto in self.tabela:
            del self.tabela[codigo_produto]
            return True
        return False

    def listar_produtos(self):
        return list(self.tabela.items())
```
